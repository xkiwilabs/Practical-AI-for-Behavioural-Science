```python
# Cell 2: Quick test — create some data and make a plot
# This checks that plotting works in your notebook.

# Create a tiny example dataset
data = pd.DataFrame({
    "Hours of sleep": [5, 6, 7, 8, 9, 6.5, 7.5, 5.5, 8.5, 7],
    "Mood rating":    [3, 5, 6, 8, 9, 4,   7,   3,   8,   6]
})

# Make a scatter plot
plt.figure(figsize=(6, 4))
sns.scatterplot(data=data, x="Hours of sleep", y="Mood rating", s=100, color="#5B8DB8")
plt.title("Does more sleep = better mood?")
plt.xlabel("Hours of sleep per night")
plt.ylabel("Self-reported mood (1–10)")
plt.tight_layout()
plt.show()

print()
print("=" * 40)
print("  Setup complete! You're ready for")
print("  PSYC4411. See you in Week 2!")
print("=" * 40)
```
